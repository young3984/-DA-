```python
import numpy as np
```

numpy 불러오기

```python
points=np.arange(-5,5,0.01)
```

-5부터 5전까지 0.01씩 증가하는 배열 생성

```python
xs,ys=np.meshgrid(points,points)
```

```python
ys
```

```python
xs
```

meshgrid는 (x,y)이 모든 쌍에 해당하는 두개의 2D행렬 생성

```python
z=np.sqrt(xs**2+ys**2)
```

```python
z
```

xs^2+ys^2의 제곱근

```python
import matplotlib.pyplot as plt
```

matplotlib.pyplot 불러오기

```python
plt.imshow(z)
```

imshow는 그려주는 함수

```python
plt.imshow(z,cmap=plt.cm.gray)
```

```python
plt.imshow(z,cmap=plt.cm.Blues)
```

색깔 조합을 선택해서 바꿀 수도 있다.

```python
plt.imshow(z,cmap=plt.cm.autumn);
plt.colorbar()
```

plt.colorbar()는 컬러바를 생성해준다

```python
plt.imshow(z,cmap=plt.cm.Blues)
plt.colorbar()
plt.title("image plot of $\sqrt{x^2 + y^2}$ for grid of value")
```

plt.title을 이용해서 제목을 그릴 수 도 있다.
